## 3) How are purchases distributed to customers?

```python
purchase_distribution = df_feb17['Телефон_new'].value_counts()

limited_distribution = purchase_distribution[purchase_distribution < 100]

plt.figure(figsize=(10, 6))
plt.hist(limited_distribution, bins=50, edgecolor='black', alpha=0.7)
plt.title('Distribution of Purchases per Customer (Limited)', fontsize=14)
plt.xlabel('Number of Purchases', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(limited_distribution, bins=50, edgecolor='black', alpha=0.7, log=True)
plt.title('Distribution of Purchases per Customer (Limited)', fontsize=14)
plt.xlabel('Number of Purchases', fontsize=12)
plt.ylabel('Number of Customers (Log Scale)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## 4) What is the average time between purchases?

```python
df_feb17['Дата'] = pd.to_datetime(df_feb17['Дата'], errors='coerce')

df_feb17 = df_feb17.dropna(subset=['Дата'])

df_feb17 = df_feb17.sort_values(by=['Телефон_new', 'Дата'])

df_feb17['time_diff'] = df_feb17.groupby('Телефон_new')['Дата'].diff().dt.days

time_differences = df_feb17['time_diff'].dropna()

time_diff_summary = time_differences.describe()

plt.figure(figsize=(10, 6))
plt.hist(time_differences, bins=50, edgecolor='black', alpha=0.7)
plt.title('Distribution of Time Between Purchases', fontsize=14)
plt.xlabel('Time Between Purchases (days)', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
files = [
    "other_data/!01_02_17_VSE.csv",
    "other_data/!03&04_17_VSE.csv",
    "other_data/!05&06_17_VSE.csv", 
    "other_data/!07&08_17_VSE.csv", 
]

all_data = []

for file in files:

    print(f"Processing file: {file}")
    temp = pd.read_csv(file, sep=';', encoding='cp1251', decimal=',', low_memory=True, on_bad_lines='skip')

    temp['ДатаЗаказаНаСайте'] = pd.to_datetime(temp['ДатаЗаказаНаСайте'], format='%d.%m.%Y %H:%M', errors='coerce')

    temp = temp[temp['ДатаЗаказаНаСайте'] >= '2017-01-01']
    temp = temp[temp['ДатаЗаказаНаСайте'] < '2017-07-01']

    temp = temp[temp['Статус'] == 'Доставлен']
    
    temp = temp.drop_duplicates()
    
    all_data.append(temp)
```

```python
final_data = pd.concat(all_data, ignore_index=True)
final_data = final_data.drop_duplicates()

final_data = final_data[final_data['Телефон_new'].notna() & 
                        (final_data['Телефон_new'] != '0') & 
                        (final_data['Телефон_new'] != ' ') & 
                        (final_data['Телефон_new'].str.contains('-'))]

final_data['quarter'] = final_data['ДатаЗаказаНаСайте'].dt.to_period('Q')

final_data
```

```python
final_data['МетодДоставки'].value_counts().plot(kind='bar')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].hist(final_data['СуммаДокумента'], bins=10, color='blue', alpha=0.7, edgecolor='black')
axes[0].set_title("Распределение сумм заказов")
axes[0].set_xlabel("Сумма заказа")
axes[0].set_ylabel("Частота")

axes[1].hist(final_data['Цена'], bins=10, color='green', alpha=0.7, edgecolor='black')
axes[1].set_title("Распределение цен SKU")
axes[1].set_xlabel("Цена SKU")
axes[1].set_ylabel("Частота")

axes[2].hist(final_data['Количество'], bins=5, color='red', alpha=0.7, edgecolor='black')
axes[2].set_title("Распределение количества позиций в заказе")
axes[2].set_xlabel("Количество товаров в заказе")
axes[2].set_ylabel("Частота")

plt.tight_layout()
plt.show()
```

```python
final_data['СуммаДокумента'] = final_data['СуммаДокумента'].astype(str).str.replace(' ', '').str.replace(',', '.').astype(float)
final_data['Количество'] = final_data['Количество'].astype(str).str.replace(' ', '').str.replace(',', '.').astype(float)

final_data['AIV'] = final_data['СуммаДокумента'] / final_data['Количество']

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# 1) Линейный график распределения цены заказа
order_price_distribution = final_data['СуммаДокумента'].value_counts().sort_index()
axes[0].plot(order_price_distribution.index, order_price_distribution.values, marker='o', linestyle='-', color='blue')
axes[0].set_title("Распределение цены заказа")
axes[0].set_xlabel("Сумма заказа")
axes[0].set_ylabel("Частота")
axes[0].grid(True)

# 2) Линейный график распределения AIV (Average Order Value)
aiv_distribution = final_data['AIV'].value_counts().sort_index()
axes[1].plot(aiv_distribution.index, aiv_distribution.values, marker='o', linestyle='-', color='green')
axes[1].set_title("Распределение AIV (Средний чек)")
axes[1].set_xlabel("Средний чек")
axes[1].set_ylabel("Частота")
axes[1].grid(True)

# Отображаем графики
plt.tight_layout()
plt.show()
```
